# Análisis de Sentimientos Periodico Página Siete 

```python
# Librerias
import pandas as pd
import json
from datetime import datetime
```

```python
f = open('./data/paginaSiete.json')
data = json.load(f)
title, date, hour = [],[],[]
for i in data:
    for r in i["titulo"]:
        title.append(r)
    for s in i["fecha"]:
        date.append(s)
    for t in i["hora"]:
        hour.append(t)
```

```python
for i in range(len(hour)):
    datetime.strptime(hour[i],'%H:%M').time()
```
