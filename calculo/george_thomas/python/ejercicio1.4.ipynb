## Elección del tamaño de la ventana de visualización

```python
# librerias
import matplotlib.pyplot as plt 
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import pylab as pl
import math
sns.set()
```

### 1.

```python
def f1(x1,x2,y1,y2):
    x = np.arange(x1,x2,0.01)
    f = x**4 - 7*x**2 + 6*x
    c,ax = plt.subplots()
    ax.plot(x,f)
    plt.ylim(y1,y2)
    ax.set(xlabel="x",ylabel="f(x)",title="$x^4 - 7x^2 + 6x$")
    plt.show()
```

```python
# a)
f1(-1,1,-1,1)
```

```python
#b
f1(-2,2,-5,5)
```

```python
#c)
f1(-10,10,-10,10)
```

```python
#d) Es la gráfica mas adecuada
f1(-5,5,-25,15)
```

### 2.

```python
def f2(x1,x2,y1,y2):
    x = np.arange(x1,x2,0.01)
    f = x**3 -4*x**2 -4*x +16 
    c,ax = plt.subplots()
    ax.plot(x,f)
    plt.ylim(y1,y2)
    ax.set(xlabel="x",ylabel="f(x)",title="$x^3 - 4x^2 - 4x + 16$")
    plt.show()
```

```python
# a)
f2(-1,1,-5,5)
```

```python
# b)
f2(-3,3,-10,10)
```

```python
# c)
f2(-5,5,-10,20)
```

```python
# d) Es la gráfica más optima para visualizar
f2(-20,20,-100,100)
```

### 3.

```python
def f3(x1,x2,y1,y2):
    x = np.arange(x1,x2,0.01)
    f = 5 + 12*x - x**3 
    c,ax = plt.subplots()
    ax.plot(x,f)
    plt.ylim(y1,y2)
    ax.set(xlabel="x",ylabel="f(x)",title="$5 + 12x - x^3$")
    plt.show()
```

```python
# a)
f3(-1,1,-1,1)
```

```python
# b)
f3(-5,5,-10,10)
```

```python
# c)
f3(-4,4,-20,20)
```

```python
# d) Es la gráfica mas ideal para visualizar
f3(-4,5,-15,25)
```

### 4.

```python
def f4(x1,x2,y1,y2):
    x = np.arange(x1,x2,0.01)
    f = (5 + 4*x - x**2)**(1/2) 
    c,ax = plt.subplots()
    ax.plot(x,f)
    plt.ylim(y1,y2)
    ax.set(xlabel="x",ylabel="f(x)",title="$\sqrt{5 + 4x - x^2}$")
    plt.show()
```

```python
# a)
f4(-2,2,-2,2)
```

```python
# b) tiene mejor gráfica para su optima visualización
f4(-2,6,-1,4)
```

```python
# c)
f4(-3,7,0,10)
```

```python
# d)
f4(-10,10,-10,10)
```
