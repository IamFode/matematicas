## 9.

```python
def fgh(x):
    if x == 0 or x == -1/4:
        print("x debe ser diferente a 0 o -1/4")
    else:
        f = lambda x: (x+1)**(1/2)
        g = lambda x: 1/(x+4)
        h = lambda x: 1/x
        print(f(g(h(x))))
```

```python
fgh(-1/3)
```

## 10.

```python
def fhg(x):
    if (2 - x) >= 0:
        f = lambda x: (x+2)/(3-x)
        g = lambda x: x**2/(x**2 + 1)
        h = lambda x: (2-x)**(1/2)
        return f(g(h(x)))
    else:
        print("2-x debe ser mayor o igual a 0")
```

```python
fhg(4)
```

### 21

```python
x = np.arange(-15, 15, 0.01)
y, y1, y2 = -x**2, -(x+7)**2, -(x-4)**2
fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x, y1)
ax.plot(x, y2)
ax.set(xlabel='x', ylabel='y',
       title='Desplazamiento de funciones')
ax.grid
plt.show()
```

### 22

```python
x = np.arange(-5,5,0.01)
y, y1, y2 = x**2, x**2 + 3, x**2 - 5
fig,ax = plt.subplots()
ax.plot(x,y,label="$y=x^2$")
ax.plot(x,y1,label="$y=x^2 + 3$")
ax.plot(x,y2,label="$y=x^2-5$")
ax.legend(fontsize="xx-small")
ax.set(xlabel="x", ylabel="y", title = "Desplzamiento")
ax.grid
ax.plot()
```
