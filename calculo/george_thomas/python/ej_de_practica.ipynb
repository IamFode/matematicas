## 41 al 48

```python
x = np.arange(-10,10,0.01)
f1, t1 = x, abs(x)
f2, t2 = x**2, abs(x)**2
f3, t3 = x**3, abs(x)**3
f4, t4 = x**2 + x, abs(x**2 + x)
f5, t5 = 4 - x**2, abs(4 - x**2)
f6, t6 = 1/x, 1/abs(x)
f7, t7 = np.sqrt(x), abs(np.sqrt(x)) 
f8, t8 = np.sin(x), np.sin(abs(x))
fig, ax = plt.subplots(2,4,figsize=(20,8))
ax[0,0].plot(f1,t1)
ax[0,0].plot(f2,t2)
ax[0,0].set_title("41")
ax[0,1].plot(x,f2)
ax[0,1].plot(x,t2)
ax[0,1].set_title("42")
ax[0,2].plot(x,f3)
ax[0,2].plot(x,t3)
ax[0,2].set_title("43")
ax[0,3].plot(x,f4)
ax[0,3].plot(x,t4)
ax[0,3].set_title("44")
ax[1,0].plot(x,f5)
ax[1,0].plot(x,t5)
ax[1,0].set_title("45")
ax[1,1].plot(x,f6)
ax[1,1].plot(x,t6)
ax[1,1].set_title("46")
ax[1,2].plot(x,f7)
ax[1,2].plot(x,t7)
ax[1,2].set_title("47")
ax[1,3].plot(x,f8)
ax[1,3].plot(x,t8)
ax[1,3].set_title("48")
plt.show()
```
