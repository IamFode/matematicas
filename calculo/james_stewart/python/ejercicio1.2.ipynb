```python
# importar 
import matplotlib.pyplot as plt 
import numpy as np
```

# 18

$$C(s)=13s + 900$$

```python
s = np.arange(0, 300, 0.01)
c = 13*s + 900

fig, ax = plt.subplots()
ax.plot(s, c)

ax.set(xlabel='Sillas producidas', ylabel='Costo en $',
       title='Costo en función de las sillas producidas')
ax.grid()
plt.show()
```

# 19

$$p(s)=1.05 - 0.10 s$$

```python
s = np.arange(1.05,60,0.01)
p = 1.05 + 0.1*s

fig, ax = plt.subplots()
ax.plot(s,p)

ax.set(xlabel="Profundidad en cm", ylabel="Presión del agua en $kg/cm^2$", title="Presión del agua en función de la profundidad")
ax.grid()
plt.show()
```
