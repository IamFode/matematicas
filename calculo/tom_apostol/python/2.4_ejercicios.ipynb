```python
# librarias
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import math
sns.set()
```

## 1.

```python
x = np.arange(-2,2,0.01)
f1 = 4 - x**2
fig, ax = plt.subplots()
ax.plot(x,f1)
plt.show()
```

## 2.

```python
x = np.arange(-2,2,0.01)
f = 4 - x**2
g = 8 - 2*x**2
fig, ax = plt.subplots()
ax.plot(x,f, label = "$f(x)=4 - x^2$")
ax.plot(x,g, label = "$g(x)=8-2x^2$")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title("Integrales")
plt.legend()
plt.show()
```
