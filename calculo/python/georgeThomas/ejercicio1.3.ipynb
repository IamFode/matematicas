## 31.

```python
def r31(x):
    x = x*(np.pi/180) 
    cos, sen = np.cos(x - np.pi/2), np.sin(x)
    print(cos,"=",sen)
    
r31(90)
```

## 32.

```python
def r32(x):
    x = x*(np.pi/180) 
    cos, sen = np.cos(x + np.pi/2), -np.sin(x)
    print(cos,"=",sen)

r32(90)
```

## 35.

```python
def r35(A, B):
    a = np.cos(A-B)
    b = np.cos(A)*np.cos(B) + np.sin(A)*np.sin(B)
    print(a, "=", b)
    
r35(np.pi, 2*np.pi)
```

$$f(x) = A \, \sin \left(\dfrac{2\pi}{B}(x - C)\right)+D$$

```python
def f(A,B,C,D,i,f):
    x = np.arange(i,f,0.001)
    f = A * np.sin((2*np.pi/B)*(x - C)) + D
    c,ax = plt.subplots()
    ax.plot(x,f)
    plt.show()
```
