<h2>Teorema 1.38</h2>
<p>Si $a \geq 0$, es $|x|\leq a$ si y sólo si $-a \leq x \leq a$</p>

```python
def teo1_38(a):
    if a >= 0:
        return '|x| <= {0} si y sólo si -{0} <= x <= {0}'.format(a)
    else:
        print("a debe ser mayor ó igual a 0")
teo1_38(10)
```

<h2>Teorema 1.39</h2>
<p>Para $x$ e $y \in \mathbb{R}$ se tiene </p> 
$$|x+y|\leq |x|+|y|$$

```python
def teo1_39(x,y):
    left = abs(x+y)  
    right = abs(x) + abs(y)
    print(left, " <= ",right)
teo1_39(2,1)
teo1_39(212,-543)
```

<h2>Teorema 1.41</h2>
$$\left( \sum_{k=1}^n a_k b_k\right)^2 \leq \left( \sum_{k=1}^n a_k^2 \right) \left( \sum_{k=1}^n b_k^2 \right)$$

```python
def teo1_41(x):
    control,con,cont = 0,0,0
    for a in range(1,x+1):
        control += a * a**2
        con += a
        cont += a**2
    print(control**2, " <= " , con**2 * cont**2 )
    
#ejemplos
teo1_41(0)
teo1_41(3)
teo1_41(1)
teo1_41(2)
teo1_41(4)
```
