```python
# Librerias
import matplotlib.pyplot as plt
import numpy as np
```

## 17.

### (i)

```python
x = np.arange(-5,5,0.02)
y = np.floor(x)

fig, ax = plt.subplots()
ax.plot(x,y)

ax.set(xlabel="x", ylabel="f(x)")
ax.grid()
plt.show()
```

### (ii)

```python
x = np.arange(-5,5,0.02)
y = x - np.floor(x)
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set(xlabel="x", ylabel="f(x)")
ax.grid()
plt.show()
```

### (iii)

```python
x = np.arange(0,10,0.02)
y = (x-np.floor(x))**(1/2)
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set(xlabel="x", ylabel="f(x)")
ax.grid()
plt.show()
```

### (iv)

```python
x = np.arange(0,10,0.02)
y = np.floor(x) + (x-np.floor(x))**(1/2)
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set(xlabel="x", ylabel="f(x)")
ax.grid()
plt.show()
```

### (v)

```python
x = np.arange(0.1,1,0.02)
y = np.floor(1/x)
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set(xlabel="x", ylabel="f(x)")
ax.grid()
plt.show()
```

### (vi)

```python
x = np.arange(0.1,1,0.02)
y = 1/np.floor(1/x)
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set(xlabel="x", ylabel="f(x)")
ax.grid()
plt.show()
```
